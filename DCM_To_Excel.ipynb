```python
import mdfreader
import pandas as pd
import numpy as np
import os
import re

def mask(t) :
    l = 'Other'
    maskk = list()
    maskk = ['DisblMsk','CtlMsk','DFES_Cls','DFES_DTCO']
    for i in maskk :
        if re.search (i,t) :
            l = i
    return l 


print ('this code is written for Mr. Ashish HM')

link = r'C:\Users\ldn1kor\Desktop\TML\TML 3.3 24 V\02_DCM'

file = r'allfiles.DCM'
path = link+'\\'+file

df = pd.read_csv(path,header = None)

heading = ['text']
df.columns = heading
dfc = r'C:\Users\ldn1kor\Desktop\TML\TML 3.3ltr\000_Summer_Trial\DFC.xlsx'
path = r'C:\Users\ldn1kor\Desktop\TML\TML 3.3ltr\000_Summer_Trial\DFC_DCM_out.xlsx'
dk = pd.read_excel(dfc)

heading = ['text']
df.columns = heading
ef = pd.DataFrame()
gb = pd.DataFrame(['Error'])
for z in dk.DFC :
    print ('\n',z)
    for idx, i in enumerate (df.text) :
          if re.search (z,i) :
            a = idx
            if re.search('DFC',i):
                for j in [1,2,3,4,5,6,7,8,9] :
                    li = j+a
                    stop = df.loc[li,'text']
                    lib = stop.split()
                    if re.search('END',stop) :
                        break
                    if re.search('WERT ',stop) :
                        for k in lib :
                            if k.isalnum()==0 :
                                lable = i.split()

                                gb['lable'] = lable [1]
                                gb ['type'] = mask (lable [1])
                                gb['value' ] = k
                                database= pd.concat([gb,ef], ignore_index = True)
                                ef = database
df = database[['lable', 'value','type']]
CtlMsk = df [(df['type']=='CtlMsk')]
CtlMsk ['value'] = CtlMsk ['value'].astype(float)
CtlMsk ['value'] = CtlMsk ['value'].astype(int)
DCtlMsk = CtlMsk [(CtlMsk['value']==512)]
DCtlMsk = DCtlMsk.reset_index(drop = True)
ECtlMsk = CtlMsk [(CtlMsk['value']!=512)]
ECtlMsk = ECtlMsk.reset_index(drop = True)
DisblMsk = df [(df['type']=='DisblMsk')]
DisblMsk ['value'] = DisblMsk ['value'].astype(float)
DisblMsk ['value'] = DisblMsk ['value'].astype(int)
DDisblMsk = DisblMsk [(DisblMsk['value']==32768)]
EDisblMsk = DisblMsk [(DisblMsk['value'] ==32768)]
DFES_Cls = df [(df['type']=='DFES_Cls')]
DFES_Cls ['value'] = DFES_Cls ['value'].astype(float)
DFES_Cls ['value'] = DFES_Cls ['value'].astype(int)
DFES_DTCO = df [(df['type']=='DFES_DTCO')]
DFES_DTCO = DFES_DTCO.reset_index(drop = True)
DFES_Cls = DFES_Cls.reset_index(drop = True)
EDisblMsk = EDisblMsk.reset_index(drop = True)
DDisblMsk = DDisblMsk.reset_index(drop = True)
ECtlMsk = ECtlMsk[['lable', 'value']]
DCtlMsk = DCtlMsk[['lable', 'value']]
EDisblMsk = EDisblMsk[['lable', 'value']]
DDisblMsk = DDisblMsk[['lable', 'value']]
DFES_DTCO = DFES_DTCO[['lable', 'value']]
DFES_Cls = DFES_Cls[['lable', 'value']]

with pd.ExcelWriter(path) as writer:
    df.to_excel(writer, sheet_name="All_Data")
    DCtlMsk.to_excel(writer, sheet_name="Disabled_CtlMsk")
    ECtlMsk.to_excel(writer, sheet_name="Enabled_CtlMsk")  
    DDisblMsk.to_excel(writer, sheet_name="Disabled_DisblMsk") 
    EDisblMsk.to_excel(writer, sheet_name="Enabled_DisblMsk")  
    DFES_Cls.to_excel(writer, sheet_name="Class")  
    DFES_DTCO.to_excel(writer, sheet_name="DTCO") 
    
    
print ('all data loaded please open Excel')
os.startfile(path)
```
